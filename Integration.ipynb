```
import numpy as np
import matplotlib.pyplot as plt
```

```
# to define a function for integration 
# takes as input, an array of x, and returns its integral
# to integrate a function of x; 
# integral = area under the curve, area upto the point
# delta_x = x[i+1] - x[i]
# integral[i] = integral[i-1] + delta_x * (y[i+1] + y[i]) * 0.5

def my_integral(x,y):
    N = len(x)
    integral = []
    integral.append(0)

    delta_x = x[1] - x[0]
    for i in range(N-1):
        temp1 = delta_x * (y[i+1] + y[i]) * 0.5
        temp2 = integral[i] + temp1
        integral.append(temp2)
    return integral
```

```
# Example 1
# to integrate a line

N=30
x = np.linspace(-2, 2, N)

b=2
m=1.2
y = m*x + b

integral_fx = my_integral(x,y)

plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,integral_fx, label= ' integral f(x)')
plt.grid()
plt.title('Function and its integral')
plt.legend()
```

```
# Example 2
# to integrate a quadratic function

N=51
x = np.linspace(-3, 3, N)

a=1
b=1
c=1
y= a*x**2 + b*x + c

integral_fx = my_integral(x,y)

plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,integral_fx, label= ' integral f(x)')
plt.grid()
plt.title('Function and its integral')
plt.legend()
```

```
# Example 3
# to integrate a sinusoidal function

N=51
x = np.linspace(0, 2*np.pi, N)

y= np.cos(x)

integral_fx = my_integral(x,y)

plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,integral_fx, label= ' integral f(x)')
plt.grid()
plt.title('Function and its integral')
plt.legend()
```

```
# Example 4
# to integrate the exponential function
# f(x) = exp(ax)

N=51
x = np.linspace(0, 5, N)

a = -0.8
y = np.exp(a*x)

integral_fx = my_integral(x,y)

plt.figure(figsize=(10,8))
plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,integral_fx, label= ' integral f(x)')
plt.grid()
plt.title('Function and its integral')
plt.legend()
```

```
# To have a function that generates the Gaussian Distribution 
# Mean = average = mew 
# Standard Deviation = Sigma
# for given input array x

def my_gaussian(mew,sigma,x):
    N = len(x)
    y = []

    const1 = sigma**2
    const2 = 2*np.pi
    const3 = 1/sigma/np.sqrt(const2)

    for i in range(N):
        temp1 = (x[i] - mew)**2
        temp2 = 0.5 * temp1 / const1
        temp3 = np.exp(-temp2) 
        temp4 = const3*temp3
        y.append(temp4)
    
    return y
```

```
# Example 5
# to generate a Gaussian Distribution 
# and then integrate it

N=301
x = np.linspace(-10, 10, N)

mew = -4
sigma = 2

y= my_gaussian(mew, sigma, x)

integral_fx = my_integral(x,y)

plt.figure(figsize=(8,6))
plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,integral_fx, label= ' integral f(x)')
plt.grid()
plt.xlabel('x')
plt.title('Gaussian Distribution Function and its integral')
plt.legend()
```

```
# Example 6
# to compute the area within (mew-sigma) to (mew+sigma) of a Gaussian Distribution 
# we compute the integral in the specified limits

mew = -2
sigma = 2
N=100
x = np.linspace((mew-3*sigma), (mew+3*sigma), N)

y= my_gaussian(mew, sigma, x)

integral_fx = my_integral(x,y)
required_area = integral_fx[-1] * 100

print('\n Area within the three sigma limits (in percentage) = ', required_area )
```

```
# Example 7
# to generate a Uniform Distribution, in the limits of x=(a,b)
# and then integrate it

a=0
b=4
N=100
x = np.linspace(a-2, b+2, N)
y=[]

for i in range(N):
    if x[i] > a and x[i] < b:
        temp = 1/(b-a)
    else:
        temp = 0
    y.append(temp)
 
integral_fx = my_integral(x,y)

plt.figure(figsize=(8,6))
plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,integral_fx, label= ' integral f(x)')
plt.grid()
plt.xlabel('x')
plt.title('Uniform Distribution Function and its integral')
plt.legend()
```

```
# Example 8
# to generate a Exponential Distribution
# and then integrate it

a=0.7
N=100
x = np.linspace(-2, 6, N)
y=[]

for i in range(N):
    if x[i] > 0:
        temp = (a) * np.exp( -a * x[i] )
    else:
        temp = 0
    y.append(temp)
 
integral_fx = my_integral(x,y)

plt.figure(figsize=(8,6))
plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,integral_fx, label= ' integral f(x)')
plt.grid()
plt.xlabel('x')
plt.title('Exponential Distribution Function and its integral')
plt.legend()
```
