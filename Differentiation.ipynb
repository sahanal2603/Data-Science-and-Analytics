```
import numpy as np
import matplotlib.pyplot as plt
```

```
# to define a function for differentiation
# takes as input, an array of x, and returns its derivative
# to differentiate a function of x; 
# derivative = delta(y)/delta(x)
# delta_x = x[i+1] - x[i]
# derivative = (y[i+1] - y[i]) / (delta_x)

def my_derivative(x,y):
    N = len(x)
    dy_by_dx = []
    dy_by_dx.append(0)    # setting the first value to 0

    delta_x = x[1] - x[0] 
    for i in range(N-1):
        temp = (y[i+1]-y[i])/delta_x
        dy_by_dx.append(temp)
    return dy_by_dx
```

```
# Example 1
# to differentiate a line

N=51
x = np.linspace(-1, 3, N)

b=2
m=3
y = m*x + b

dy_by_dx = my_derivative(x,y)

plt.figure(figsize=(8,6))
plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,dy_by_dx, label= ' f_dash_x', lw=5)
plt.grid()
plt.title('Function and its derivative')
plt.legend()
```

```
# Example 2
# to differentiate a quadratic function

N=51
x = np.linspace(-3, 3, N)

a=1
b=1
c=1
y= a*x**2 + b*x + c

dy_by_dx = my_derivative(x,y)

plt.figure(figsize=(10,8))
plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,dy_by_dx, label= ' f_dash_x', lw=5)
plt.grid()
plt.title('Function and its derivative')
plt.legend()
```

```
# Example 3
# to differentiate a sinusoidal function

N=51
x = np.linspace(0, 2*np.pi, N)

y= np.cos(x)

dy_by_dx=my_derivative(x,y)

plt.figure(figsize=(8,6))
plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,dy_by_dx, label= ' f_dash_x', lw=5)
plt.grid()
plt.title('Function and its derivative')
plt.legend()
```

```
# Example 4
# to differentiate an exponential function
# f(x) = exp(ax)

N=51
x = np.linspace(0, 5, N)

a = -1
y = np.exp(a*x)

dy_by_dx=my_derivative(x,y)

plt.figure(figsize=(8,6))
plt.plot(x,y, label='f(x)', lw=5)
plt.plot(x,dy_by_dx, label= ' f_dash_x')
plt.grid()
plt.title('Function and its derivative')
plt.legend()
```
