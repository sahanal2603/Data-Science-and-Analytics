```
# To compute the Covariance and Correlation of given arrays

CIE=np.array([3,   3, 4,  4,    5, 5,   6,  6,   7,   7,   8, 8,    9, 9])
SEE=np.array([95, 60, 85, 80,  75, 55,  65,  60, 55,  50,  45, 40, 35, 85    ])

# to compute the Covariance, Correlation

N=len(CIE)
mew_CIE = np.mean(CIE)
mew_SEE = np.mean(SEE)
sigma_CIE = np.std(CIE)
sigma_SEE = np.std(SEE)
mew_CIE_SEE = np.dot(CIE,SEE)/N

covariance = (mew_CIE_SEE) - (mew_CIE)*(mew_SEE)
correlation = covariance /(sigma_CIE * sigma_SEE)

print('\n Covariance of CIE and SEE = ',covariance )
print('\n Correlation of CIE and SEE = ',correlation )
```

**Some useful Functions**

Factorial of a number

```
# to write a function to compute factorial of a given number

def my_factorial(k):
    val = 1
    for i in range(1,k+1):
        val = val*i
    return val

print('Factorial = ', my_factorial(3))
```

NCK: accepts N, K, and returns N!/(K! (N-K)!)

```
# to have a function to compute nCk
# nCk = n!/(k! (n-k)!)

def NCK(N,k):
    num=1
    den=1
    A = min(k,(N-k))
    for i in range(A):
        num = num*(N-i)
        den = den*(i+1)
        
    return (num/den)
```

To Differentiate the given Function

```
# to define a function for differentiation
# takes as input, an array of x, and returns its derivative
# to differentiate a function of x; 
# derivative = delta(y)/delta(x)
# delta_x = x[i+1] - x[i]
# derivative = (y[i+1] - y[i]) / (delta_x)

def my_derivative(x,y):
    N = len(x)
    dy_by_dx = []
    dy_by_dx.append(0)    # setting the first value to 0

    delta_x = x[1] - x[0] 
    for i in range(N-1):
        temp = (y[i+1]-y[i])/delta_x
        dy_by_dx.append(temp)
    return dy_by_dx
```

Integrate a given given function

```
# to define a function for integration 
# takes as input, an array of x, and returns its integral
# to integrate a function of x; 
# integral = area under the curve, area upto the point
# delta_x = x[i+1] - x[i]
# integral[i] = integral[i-1] + delta_x * (y[i+1] + y[i]) * 0.5

def my_integral(x,y):
    N = len(x)
    integral = []
    integral.append(0)

    delta_x = x[1] - x[0]
    for i in range(N-1):
        temp1 = delta_x * (y[i+1] + y[i]) * 0.5
        temp2 = integral[i] + temp1
        integral.append(temp2)
    return integral
```

compute the Binomial distribution

```
def my_binomial_dist(N,p):
    q = (1-p)
    PK = []
    for i in range(N+1):
        C1 = NCK(N,i)
        C2 = np.power(p,i)
        C3 = np.power(q,(N-i))
        PK.append(C1*C2*C3)
    
    return PK
```

Entropy Calculator: Accepts the discrete probability distribution

```
def my_entropy(prob_dist):
    ent = 0
    N = len(prob_dist)
    for i in range(N):
        C1 = - prob_dist[i]
        C2 = np.log2(prob_dist[i])
        ent = ent + C1*C2
    
    return ent
```

**Some Continuous Distributions**

The Uniform Distribution

```
# To have a function that generates the Uniform Distribution 
# f(x) = 1/(b-a)   for a <= x  <= b, and 0 otherwise
# for given input array x

def my_uniform(a,b,x): 
    y=[]
    N=len(x)
    for i in range(N):
        if x[i] > a and x[i] < b:
            temp = 1/(b-a)
        else:
            temp = 0
        y.append(temp)
    
    return y
```

The Exponential Distribution

```
# To have a function that generates the Exponential Distribution 
# f(x) = (a) exp(-ax)   for x >=0 , and 0 otherwise
# for given input array x

def my_exponential(a,x):
    y=[]
    N=len(x)

    for i in range(N):
        if x[i] > 0:
            temp = (a) * np.exp( -a * x[i] )
        else:
            temp = 0
        y.append(temp)
    
    return y
```


The Gaussian Distribution/Normal Distribution

```
# To have a function that generates the Gaussian Distribution 
# Mean = average = mew 
# Standard Deviation = Sigma
# for given input array x

def my_gaussian(mew,sigma,x):
    N = len(x)
    y = []

    const1 = sigma**2
    const2 = 2*np.pi
    const3 = 1/sigma/np.sqrt(const2)

    for i in range(N):
        temp1 = (x[i] - mew)**2
        temp2 = 0.5 * temp1 / const1
        temp3 = np.exp(-temp2) 
        temp4 = const3*temp3
        y.append(temp4)
    
    return y
```

**Some Discrete Distributions**

The Poisson Distribution

```
# The discrete Poisson Distribution
# P(x=k)= (lamda^k) * exp(-lamda) / k! 

lmd = 6.5

# tp compute the Poisson distribution for k = 0 to 12

N = 10
poisson_k =[]

for i in range(N):
    temp = (lmd**i) * (np.exp(-lmd)) /my_factorial(i)
    poisson_k.append(temp)
    
k=np.arange(N)

plt.bar(k,poisson_k)
plt.grid()
plt.xlabel('x')
plt.ylabel('P(k)')
plt.title('Poisson Density Function')

print('\n Sum of the probabilities = ',np.sum(poisson_k))
print('\n Mean of the Possion Distribution = ',np.dot(k,poisson_k))
print('\n Entropy of the Poisson Distribution = ', my_entropy(poisson_k))
print("\n poisson_k = ", poisson_k)
print("\n poisson_k = 3", poisson_k[3])
print('\n')
```

To compute CDF (for x<<XX)

```
# to generate a Exponential Distribution
# and then integrate it

a=0.7
XX = 2
N=100
x = np.linspace(-2, XX, N)

y=my_exponential(a,x)

integral_fx = my_integral(x,y)
print('CDF for x less than XX = ', integral_fx[-1])
print('CDF for x greater than XX = ', 1-integral_fx[-1])
print(' Interesting to observe that CDF greater than x =', np.exp(-a*XX) )
```

Conditional Probability of Exponential distribution

```
# to obtain the conditional probability that
# CDF of x>2/ given CDF of x>1

a=0.7
XX = 2.8    # chose Xx greater than YY
YY = 1.3
delta_x = XX-YY
XX = YY + delta_x
CDF_x_greater_than_XX = np.exp(-a*XX)
CDF_x_greater_than_YY = np.exp(-a*YY)
CDF_x_greater_than_delta_x = np.exp(-a*delta_x)
conditional_prob = CDF_x_greater_than_XX/CDF_x_greater_than_delta_x
print(' CDF x greater than XX =', CDF_x_greater_than_XX )
print(' CDF x greater than YY =', CDF_x_greater_than_YY )
print('Conditional probability = ', conditional_prob)
```
