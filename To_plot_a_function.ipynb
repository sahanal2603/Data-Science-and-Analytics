```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Example 1
# To plot y=sin(x)
# np.linspace(start, end, Number of values)


x=np.linspace(0,2*np.pi,21)  # 0 to 2 pi values, 21 values
y=np.sin(x)

plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y=sin(x)')
plt.title('Sin(x)')
```

```
# Example 2
# To plot y=sin(x), cos(x)

x=np.linspace(0,4*np.pi,50)
y1=np.sin(x)
y2=np.cos(x)

# c for Colour, lw for line width(thickness)
plt.plot(x,y1, label='sin(x)', c='g', lw=4 )
plt.plot(x,y2, label='cos(x)', c='r' )
plt.grid()
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.title('Trigonometric Functions')
plt.legend()
```

```
# Example 3
# To plot y=sin(x), sin^2(x)

x=np.linspace(0,4*np.pi,50)
y1=np.sin(x)
y2=y1**2

#When sin function is squared, the frequency doubles
plt.plot(x,y1, label='sin(x)',c='b', lw=3 )
plt.plot(x,y2, label='sin^2(x)', c='g', lw=3)
plt.grid()
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.title('Trigonometric Functions')
plt.legend()
```

```
# Example 4
# To plot the line y=mx+b

b=-5
m=2

x=np.linspace(-5,5,30)
N=len(x)

y = m*x + b

plt.plot(x,y)
plt.grid()
plt.axis([-5,5, -10,10])
plt.xlabel('x')
plt.ylabel('y=mx+b')
```

```
# Example 5
# To plot the line y=ax^2 + bx +c

a=-1
b=0
c=0

x=np.linspace(-5,5,30)
N=len(x)

y = a*(x**2) + b*x + c

plt.plot(x,y)
plt.grid()
plt.axis([-5,5, -10,10])
plt.xlabel('x')
plt.ylabel('y=ax^2 + bx +c')
```

```
# Example 6
# To plot y=exp(-ax) for x >0;

N=6
x=np.linspace(0, 5, N)
a=1
y=[]
for i in range(N):
    temp = np.exp(-a*x[i])
    y.append(temp)
    
plt.plot(x,y, 'x-')
plt.grid()
plt.xlabel('x')
plt.ylabel('exp(-x)')
```

```
# Example 7
# To plot y=exp(-ax) for x >0; and zero otherwise

N=51
x=np.linspace(-2, 5, N)
a=1
y=[]
for i in range(N):
    if x[i] < 0:
      temp=0
    else: 
      temp = np.exp(-a*x[i])
      
    y.append(temp)

plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('exp(-x)')
```

```
# Example 8
# To plot y=(1/sqrt(2pi))exp(-(x)^2/2)
# Normal distribution with zero mean, and Unit Standard Distribution

N=101
x=np.linspace(-3, 3, N)

fx=[]
for i in range(N):
    temp1=1/np.sqrt(2*np.pi)
    temp2=x[i]**2
    temp3=np.exp(-temp2)
    temp4=temp1*temp3
    fx.append(temp4)

plt.plot(x,fx)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Normal Distribution')
```

```
# Example 9
# To plot y=(1/sqrt(2pi sigma^2))exp(-(x-mew)^2/2/sigma^2)
# Normal distribution with zero mean, and Unit Standard Distribution

N=101
x=np.linspace(-12, 12, N)
mew=3
sigma=0.33

fx=[]
const1=(sigma**2)
const2=(2*np.pi)
const3=np.sqrt(const1*const2)
const4=1/const3

for i in range(N):
    temp1 = (x[i]-mew)**2
    temp2 = np.exp(-temp1/2/const1)
    temp3 = const4 * temp2
    fx.append(temp3)

plt.plot(x,fx)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Normal Distribution')
```

```
# Example 10
# to plot the sigmoid function
# to plot g(z) = 1/(1+exp(-z))
N=101
z=np.linspace(-8, 8, N)

gz=[]

for i in range(N):
    temp1 = (1+np.exp(-z[i]))
    gz.append(1/temp1)
    
plt.plot(z,gz)
plt.grid()
plt.xlabel('z')
plt.ylabel('g(z)')
plt.title('Sigmoid Function')
```

```
# Example 11a - Homework
# To plot the function f(x) = (1/a) exp(-ax) or  y=(1/a)exp(-ax), for a = 1,2,5
# a = 1

N = 50
x=np.linspace(-2, 5, N)
a=1
y=[]

for i in range(N):
    temp = (1/a)*np.exp(-a*x[i])
    y.append(temp)

plt.plot(x,y, 'x-')
plt.grid()
plt.xlabel('x')
plt.ylabel('(1/a) exp(-ax)')

```
