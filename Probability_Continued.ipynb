```
import numpy as np
import matplotlib.pyplot as plt
```

```
# to write a function to compute factorial of a given number

def my_factorial(k):
    val = 1
    for i in range(1,k+1):
        val = val*i
    return val

print('Factorial = ', my_factorial(4))
```

**To generate the Binomial Distribution**

```
# number of favourable events in a set of independent 
# binary events. Yes with probability = p, and No with probability =(1-p)
# to compute the probability of number of Heads p(k)


p = 0.25    # probability of Head
q = (1-p)   # probability of NOT Head 

N = 9      # number of times coin tossed

PK = []

const=my_factorial(N)

for k in range(N+1):
    temp1 = my_factorial(k)
    temp2 = my_factorial(N-k)
    k_c_N = const/(temp1*temp2)
    temp3 = np.power(p,k)
    temp4 = np.power(q,(N-k))
    prob = k_c_N * temp3 * temp4
    PK.append(prob)

n = np.arange(N+1)
plt.bar(n,PK)
plt.xlabel('k (Number of favourable events) ')
plt.ylabel('Probability - P(k)')
plt.title('Binomial distribution')
plt.grid()

actual_mean = np.dot(n,PK)
n_new= n - actual_mean
squared_n = np.power(n_new,2)
actual_variance = np.dot(squared_n, PK)
print('\n Acutal Mean =          ', round(actual_mean,3))
print('\n Compare with product Np = ', N*p)
print('\n Actual Variance =     ', round(actual_variance,3) )
print('\n Compare with product Npq = ', N*p*q)
print('\n')
```

To generate the Poisson Distribution

```
# to write a function to compute factorial of a given number

def my_factorial(k):
    val = 1
    for i in range(1,k+1):
        val = val*i
    return val

print('Factorial = ', my_factorial(3))
```

```
# The discrete Poisson Distribution
# P(x=k)= (lamda^k) * exp(-lamda) / k! 

lmd = 9.5

# tp compute the Poisson distribution for k = 0 to 12

N = 15
poisson_k =[]

for i in range(N):
    temp = (lmd**i) * (np.exp(-lmd)) /my_factorial(i)
    poisson_k.append(temp)
    
k=np.arange(N)

plt.scatter(k,poisson_k)
plt.grid()
plt.xlabel('x')
plt.ylabel('P(k)')
plt.title('Poisson Density Function')

print('\n Sum of the probabilities = ',np.sum(poisson_k))
print('\n')
print(np.dot(k,poisson_k))

print( 'Number of Arrivals        Proability')
for i in range(N):
    print('\t',i, ' \t \t ', round(poisson_k[i],2))

print('\n')
```

To integrate a given function f(x)



```
# to define a function for integration 
# takes as input, an array of x, and returns its integral
# to integrate a function of x; 
# integral = area under the curve, area upto the point
# delta_x = x[i+1] - x[i]
# integral[i] = integral[i-1] + delta_x * (y[i+1] + y[i]) * 0.5

def my_integral(x,y):
    N = len(x)
    integral = []
    integral.append(0)

    delta_x = x[1] - x[0]
    for i in range(N-1):
        temp1 = delta_x * (y[i+1] + y[i]) * 0.5
        temp2 = integral[i] + temp1
        integral.append(temp2)
    return integral
```

**The Central Limit Theorem**

The joint probability distribution of identical independent distributions, is given by the convolution of the individual distributions. For N large, the joint probability distribution tends to Gaussian Distribution

Consider, N identical independent Uniform distribution

```
# To have a function that generates the Uniform Distribution 
# f(x) = 1/(b-a)   for a <= x  <= b, and 0 otherwise
# for given input array x

def my_uniform(a,b,x): 
    y=[]
    N=len(x)
    for i in range(N):
        if x[i] > a and x[i] < b:
            temp = 1/(b-a)
        else:
            temp = 0
        y.append(temp)
    
    return y

a = -2
b = 2
N=45
x=np.linspace(-3,3,N)
y1 = my_uniform(a,b,x)      # The probability Distribution Function
y2 = np.convolve(y1, y1)    # The joint Probability for TWO 
y3 = np.convolve(y2,y1)     # The joint Probability for THREE 
y4 = np.convolve(y3,y1)     # The joint Probability for FOUR

plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(y1)
plt.title('The Probability Density Function')
plt.subplot(2,2,2)
plt.plot(y2)
plt.title('joint probability: of TWO variable ')
plt.subplot(2,2,3)
plt.plot(y3)
plt.title('joint probability: of THREE variable ')
plt.subplot(2,2,4)
plt.plot(y4)
plt.title('joint probability: of FOUR variable ')
```

To generate the Exponential distribution, given a and x



```
# To have a function that generates the Exponential Distribution 
# f(x) = (a) exp(-ax)   for x >=0 , and 0 otherwise
# for given input array x

def my_exponential(a,x):
    y=[]
    N=len(x)

    for i in range(N):
        if x[i] > 0:
            temp = (a) * np.exp( -a * x[i] )
        else:
            temp = 0
        y.append(temp)
    
    return y


a = 0.8
N = 300
lower_limit = 0
upper_limit = 4 * (1/a)
x = np.linspace(lower_limit, upper_limit, N)
y1 = my_exponential(a, x)
y2 = np.convolve(y1,y1)
y3 = np.convolve(y2,y1)
y4 = np.convolve(y3,y1)
y5 = np.convolve(y4,y1)
y6 = np.convolve(y5,y1)
y7 = np.convolve(y6,y1)
y8 = np.convolve(y7,y1)
y9 = np.convolve(y8,y1)

plt.figure(figsize=(15,10))
plt.subplot(2,3,1)
plt.plot(y1)
plt.title('joint probability density function ')
plt.subplot(2,3,2)
plt.plot(y2)
plt.title('joint probability: of TWO variables ')
plt.subplot(2,3,3)
plt.plot(y3)
plt.title('joint probability: of THREE variables ')
plt.subplot(2,3,4)
plt.plot(y4)
plt.title('joint probability: of FOUR variables ')
plt.subplot(2,3,5)
plt.plot(y5)
plt.title('joint probability: of FIVE variables ')
plt.subplot(2,3,6)
plt.plot(y6)
plt.title('joint probability: of SIX variables ')
plt.subplot(2,3,7)
plt.plot(y7)
plt.title('joint probability: of SEVEN variables ')
plt.subplot(2,3,8)
plt.plot(y8)
plt.title('joint probability: of EIGHT variables ')
plt.subplot(2,3,9)
plt.plot(y9)
plt.title('joint probability: of NINE variables ')
```

```
# To have a function that generates the Gaussian Distribution 
# Mean = average = mew 
# Standard Deviation = Sigma
# for given input array x

def my_gaussian(mew,sigma,x):
    N = len(x)
    y = []

    const1 = sigma**2
    const2 = 2*np.pi
    const3 = 1/sigma/np.sqrt(const2)

    for i in range(N):
        temp1 = (x[i] - mew)**2
        temp2 = 0.5 * temp1 / const1
        temp3 = np.exp(-temp2) 
        temp4 = const3*temp3
        y.append(temp4)
    
    return y

mew = 2
sigma = 1
lower_limit = mew  - 3*sigma
upper_limit = mew + 3*sigma
N = 300
x = np.linspace(lower_limit, upper_limit, N)
y1 = my_gaussian(mew, sigma, x)
y2 = np.convolve(y1, y1)
y3 = np.convolve(y2,y1)
y4 = np.convolve(y3,y1)

plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(y1)
plt.title('The Probability Density Function')
plt.subplot(2,2,2)
plt.plot(y2)
plt.title('joint probability: of TWO variable ')
plt.subplot(2,2,3)
plt.plot(y3)
plt.title('joint probability: of THREE variable ')
plt.subplot(2,2,4)
plt.plot(y4)
plt.title('joint probability: of FOUR variable ')
```

**Population and Sample Statitics**

```
# let us consider a population of 
# size N=5000 with mean= 400 and standard deviation = 90

N=5000
mew = 400
sigma = 120
noise_gaussian = np.random.randn(N) 
our_population=noise_gaussian*sigma + mew

plt.hist(our_population)
```

```
# we will take sample from the population

N_sample = 400

sample = np.random.choice(our_population,N_sample)

print('\n Mean of Sample 1  = ', np.mean(sample)) 
print('\n Standard Deviation of Sample 1 = ', np.std(sample))
print('\n')

plt.hist(sample)
```
